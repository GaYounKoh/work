## Q4.
Target(따릉이 대여량)을 EDA 했을 때, 대여량이 현저히 적은 날이 있었다. <br>
그 날(따릉이 대여량 <= 8000)은 대여량이 많은 날에 비하여 Feature에서 어떤 특징이 띄었는가?

- 강수여부(형태)에서 target과 반비례 관계를 보였다.
- 풍속과 target 사이에 반비례 관계가 보였다.
- 최저 기온과 target 사이에 비례 관계가 보였다.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family = 'NanumBarunGothic')
import seaborn as sns

train = pd.read_csv('따릉이 data/train.csv')
# train.info()
```

```python
x = train[train['number_of_rentals']<=8000].drop(['date_time', 'number_of_rentals'], axis = 1).mean().index
y = train[train['number_of_rentals']<=8000].drop(['date_time', 'number_of_rentals'], axis = 1).mean().values

plt.figure(figsize = (18,5), dpi=100)

plt.xlabel("피쳐")
plt.ylabel("수치")
plt.bar(x, y, label = '이하')


x = train[train['number_of_rentals']>8000].drop(['date_time', 'number_of_rentals'], axis = 1).mean().index
y = train[train['number_of_rentals']>8000].drop(['date_time', 'number_of_rentals'], axis = 1).mean().values

plt.title("대여량 8000 기준 피쳐들의 평균 값 비교")
plt.bar(x, y, alpha = 0.5, label = '초과')
plt.legend()
plt.show()
```

```python
_ = plt.figure(figsize= (15,5), dpi = 100)
_ = sns.regplot(x='precipitation_form', y='number_of_rentals', data=train)
_ = plt.xlabel('강수 여부')
_ = plt.ylabel('따릉이 대여량')

# 강수 여부와 따릉이 대여량 사이에 반비례 관계가 보임.
```

```python
plt.figure(figsize= (15,5))
sns.regplot(x='wind_speed', y='number_of_rentals', data=train)
_ = plt.xlabel('풍속')
_ = plt.ylabel('따릉이 대여량')

# 풍속과 따릉이 대여량 사이에 반비례 관계가 보임.
```

```python
plt.figure(figsize= (15,5))
sns.regplot(x='low_temp', y='number_of_rentals', data=train)
_ = plt.xlabel('최저 기온')
_ = plt.ylabel('따릉이 대여량')

# 최저 기온과 따릉이 대여량 사이에 비례관계가 보임.
```

```python
def compare(idx):
  x = ['8000<=' + low.index[idx], '8000>' + high.index[idx]]
  y = [low.values[idx], high.values[idx]]

  plt.figure(figsize = (8,5))
  plt.title(low.index[idx]+ " 의 평균")
  plt.xlabel('피쳐')
  plt.ylabel('수치')
  plt.bar(x, y)
  plt.show()

low = train[train['number_of_rentals']<=8000].mean()
high = train[train['number_of_rentals']>8000].mean()
```
