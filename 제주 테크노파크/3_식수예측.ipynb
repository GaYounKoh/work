#### library import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

#### data load

```python
train = pd.read_csv('/project/kelly/2022 여름 데이콘 교육/235743_구내식당 식사 인원 예측 AI 경진대회_data/train.csv')
test = pd.read_csv('/project/kelly/2022 여름 데이콘 교육/235743_구내식당 식사 인원 예측 AI 경진대회_data/test.csv')
```

## 중식계

### 중식계 이상치 발생 확인

```python
plt.figure(dpi=150)
plt.title("일별 중식계")
plt.xlabel('일')
plt.ylabel('중식계')
plt.plot(train.index, train['중식계'])
plt.hlines(y=500, xmin=0, xmax=1230, color='red', linestyle='dotted')
plt.show()
```

### 상자그림을 통한 이상치(범위) 확인
(수염 밖의 o이 이상치(outlier))

```python
_ = plt.boxplot(train['중식계'])
```

### 이상치 부분 데이터 살펴보고 이상치 발생 원인 생각하기

```python
train[train['중식계'] < 500].head()

# 상자그림에서는 400 미만 또는 1400 이상이 이상치라고 하고있으나
# 교육시간에는 500 미만을 이상치라고 설정.

# 대체로 본사휴가자수가 많은 날 (주로 공휴일 전후 주말 사이에 끼인 날)이 추출됨을 확인.
```

###  중식계 대비 이상치에 가장 영향이 있을법한 본사휴가자수 데이터 비교 플랏.

```python
x = ['중식<500인 날', '중식>=500인 날']
y = [train[train['중식계'] < 500]['본사휴가자수'].mean(), train[train['중식계'] >= 500]['본사휴가자수'].mean()]

plt.figure(dpi=150)
plt.title("중식계 vs 휴가자수")
plt.xlabel('중식계')
plt.ylabel('휴가자수')
plt.bar(x, y)
plt.show()
```

## 석식계

### 석식계 이상치 발생 확인

```python
plt.figure(dpi=150)
plt.title("일별 석식계")
plt.xlabel('일')
plt.ylabel('석식계')
plt.plot(train.index, train['석식계'], alpha=0.8)

plt.hlines(y=50, xmin=0, xmax=1230, color='red', linestyle='dotted')

# 3월
plt.vlines(x=980, ymin=0, ymax=900, color='pink', linestyle='-.')
# 7월
plt.vlines(x=1080, ymin=0, ymax=900, color='pink', linestyle='-.')

_ = plt.annotate(text = f'3-7월', xy = (1078, 850), xytext = (1150, 869),
               arrowprops = dict(arrowstyle = '->', facecolor = 'red'),
                 fontsize = 10)
# plt.text(1098, 839, '3-7월')
# plt.plot([980,980,1080,1080,980], [900,0,0,900,900], color='pink', linestyle='-.')
plt.fill_between([980,980,1080,1080,980], [900,0,0,900,900], color='pink', alpha = 0.5)
plt.show()
```

### 상자그림을 통한 이상치(범위) 확인
(수염 밖의 o이 이상치(outlier))

```python
_ = plt.boxplot(train['석식계'])
```

### 이상치 부분 데이터 살펴보고 이상치 발생 원인 생각하기

```python
# print(train[train['석식계'] == 0]['요일'].unique())
print(train[train['석식계'] == 0]['요일'].value_counts())
train[train['석식계'] == 0].head(10)

# 박스플랏은 200 이하, 700 초과 부분을 이상치라고 하고있으나,
# 교육시간에는 0인 것만 이상치로 설정.

# 주로 수요일이 추출됨을 확인. 자기 계발의 날으로 특정된 경우도 있으나 특정되지 않은 날도 있었음.
```

### 요일별 석식계 분포

```python
train['요일'] = train['요일'].map({'월':1, '화':2, '수':3, '목':4, '금':5})
```

```python
x = ['월', '화', '수', '목', '금']
y = train.groupby('요일').mean()['석식계'].sort_index().values

plt.figure(dpi=150)
plt.title("요일 별 석식계")
plt.xlabel('요일')
plt.ylabel('석식계')
plt.bar(x, y)
plt.show()
```

### 월별 석식계 자세히 보기

```python
def month(text:str):
    return text[5:7]

train['월'] = train['일자'].map(month)
# _ = plt.figure(dpi = 150)
train.groupby('월').mean()[['석식계']].plot(figsize=(15,5))
# plt.vlines(x=2, ymin=350, ymax=600, color='red', linestyle='dotted')
# plt.vlines(x=6, ymin=350, ymax=600, color='green', linestyle='dotted')
plt.ylim(350, 600)
plt.fill_between([2,2,6,6,2], [600,350,350,600,600], color='pink', alpha = 0.5)
plt.title('3월-7월 석식계 추이', fontsize = 20)
plt.show()

# 3-7월 구간은 감소 경향을 보임.
```

```python
def month(text:str):
    return text[5:7]

train['월'] = train['일자'].map(month)
train.groupby('월').mean()[['석식계']].plot(figsize=(15,5), marker = 'o')
# # 3월
# plt.vlines(x=2, ymin=350, ymax=600, color='red', linestyle='dotted')
# # 10월
# plt.vlines(x=9, ymin=350, ymax=600, color='green', linestyle='dotted')

plt.ylim(350, 600)
_ = plt.annotate(text = f'3월', xy = (2, 532), xytext = (2.2, 550),
               arrowprops = dict(arrowstyle = '->', facecolor = 'red'),
                 fontsize = 10)

_ = plt.annotate(text = f'10월', xy = (9, 482), xytext = (9.2, 500),
               arrowprops = dict(arrowstyle = '->', facecolor = 'red'),
                 fontsize = 10)

plt.fill_between([2,2,9,9,2], [600,350,350,600,600], color='pink', alpha = 0.5)
plt.title('월별 석식계 추이 (3-10월)', fontsize = 20)

plt.show()

# 3, 10월은 다른 달에 비해 상대적으로 석식계가 높음.
```

### 연월별 석식계 자세히 보기

```python
def Y_month(text:str):
    return text[:7]
train['월'] = train['일자'].map(Y_month)
train.groupby('월').mean()[['석식계']].plot(figsize=(15,5), marker = 'o', markersize = 3.5,)

# #3월, 10월
x = np.array([1,1,8,8,1])

plt.fill_between(x, [600,330,330,600,600], color='pink', alpha = 0.5)
plt.fill_between(x+12, [600,330,330,600,600], color='pink', alpha = 0.5)
plt.fill_between(x+24, [600,330,330,600,600], color='pink', alpha = 0.5)
plt.fill_between(x+36, [600,330,330,600,600], color='pink', alpha = 0.5)
plt.fill_between(x+48, [600,330,330,600,600], color='pink', alpha = 0.5)
plt.title('연월별 석식계 추이', fontsize = 20)

plt.ylim(330, 600)
plt.show()

# 전체 연도에서도 3, 10월이 다른 달에 비해 상대적으로 석식계가 높음.
```

### 일별 석식계 자세히 보기

```python
def day(text:str):
    return text[8:]

train['일'] = train['일자'].map(day)
train.groupby('일').mean()[['석식계']].plot(figsize=(15,5))
plt.title("일별 석식계")
plt.show()

# 석식계 월초 가장 많고, 월말까지 지속적 감소함을 확인
```
